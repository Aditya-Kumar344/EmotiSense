```python
X_train = train_df.drop("species", axis=1)
y_train = train_df["species"]
X_test = test_df.drop("species", axis=1)
y_test = test_df["species"]
```

```python
y_train[y_train=="Iris-versicolor"] = 0
y_train[y_train=="Iris-virginica"] = 1
y_test[y_test=="Iris-versicolor"] = 0
y_test[y_test=="Iris-virginica"] = 1
```

```python
X_train = np.array(X_train, dtype=np.float64)
y_train = np.array(y_train, dtype=np.float64)
X_test = np.array(X_test, dtype=np.float64)
y_test = np.array(y_test, dtype=np.float64)
```

```python
def sigmoid(z):
 return 1/(1+np.exp(-z))
```

```python
def calculate_cost(X, y, W, b):
  m,n = X.shape
  loss_sum = 0
  for i in range(m):
    z = 0
    for j in range(n):
      z += W[j] * X[i][j]
    z += b
    y_hat = sigmoid(z)
    loss = -(y[i] * np.log(y_hat)) - ((1-y[i]) * np.log(1-y_hat))
    loss_sum += loss
  cost = loss_sum/m
  return cost
```

```python
def gradient(X, y, W, b):
  m,n = X.shape
  dj_dw = np.zeros(W.shape)
  dj_db = 0
  for i in range(m):
    z = 0
    for j in range(n):
      z += W[j] * X[i][j]
    z += b
    y_hat = sigmoid(z)
    dj_db += (y_hat - y[i])
    for j in range(n):
      dj_dw[j] += (y_hat - y[i]) * X[i][j]
  dj_db /= m
  dj_dw /= m
  return dj_db, dj_dw
```

```python
def GD(X, y, W, b, lr, iters):
  m,n = X.shape
  costs = []
  for i in range(iters):
    dj_db, dj_dw = gradient(X, y, W, b)
    W -= lr * dj_dw
    b -= lr * dj_db
    cost = calculate_cost(X, y, W, b)
    costs.append(cost)
    if (i%500==0 or i == iters-1):
      print(f"Iteration {i}: Cost = {cost:8f}")
  return W, b, costs
```

```python
W_in = np.zeros(X_train.shape[1])
b_in = 0
W, b, cost_history = GD(X_train, y_train, W_in, b_in, 0.1, 10000)
```

```python
def predict(X, W, b):
  m, n = X.shape
  y_pred = np.zeros(m)
  for i in range(m):
    z = 0
    for j in range(n):
      z += W[j] * X[i][j]
    z += b
    y_hat = sigmoid(z)
    y_pred[i] = round(y_hat)
  return y_pred
```
